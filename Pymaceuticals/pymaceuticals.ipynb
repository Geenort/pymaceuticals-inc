```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
data_df = pd.merge(clinical_trial_df, mouse_drug_df, on="Mouse ID")

# Display the data table for preview
data_df.head()

```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
MTV_grp = data_df[["Timepoint", "Tumor Volume (mm3)","Drug"]].groupby(["Drug","Timepoint"])

# Convert to DataFrame
MTV_df = MTV_grp.mean().reset_index()

# Preview DataFrame
MTV_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame
SETV_df = MTV_grp.sem().reset_index()
# Preview DataFrame
SETV_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_df = MTV_grp.mean().unstack(level=0)
tumor_df.columns = tumor_df.columns.droplevel(0)

se_tumor_df = MTV_grp.sem().unstack(level=0)
se_tumor_df.columns = se_tumor_df.columns.droplevel(0)

# Preview that Reformatting worked
tumor_df.head()
```

```python
# Generate the Plot (with Error Bars)
capomulin = plt.errorbar(tumor_df.index,
                         tumor_df["Capomulin"],
                         yerr=se_tumor_df["Capomulin"],
                         color='r',
                         marker='o',
                         linestyle='--')
infubinol = plt.errorbar(tumor_df.index,
                         tumor_df["Infubinol"],
                         yerr=se_tumor_df["Infubinol"],
                         color='b',
                         marker='^',
                         linestyle='--')
ketapril = plt.errorbar(tumor_df.index,
                         tumor_df["Ketapril"],
                         yerr=se_tumor_df["Ketapril"],
                         color='green',
                         marker='s',
                         linestyle='--')
placebo = plt.errorbar(tumor_df.index,
                         tumor_df["Placebo"],
                         yerr=se_tumor_df["Placebo"],
                         color='black',
                         marker='d',
                         linestyle='--')

plt.grid()
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(handles=[capomulin,infubinol,ketapril,placebo])

# Save the Figure
plt.savefig("images/tumor_response_to_treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
MMS_grp = data_df[["Timepoint","Metastatic Sites","Drug"]].groupby(["Drug","Timepoint"])

# Convert to DataFrame
MMS_df = MMS_grp.mean()

# Preview DataFrame
MMS_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

# Convert to DataFrame
SEMS_df = MMS_grp.sem()

# Preview DataFrame
SEMS_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_df = MMS_df.unstack(level=0)
metastatic_df.columns = metastatic_df.columns.droplevel(0)

se_metastatic_df = SEMS_df.unstack(level=0)
se_metastatic_df.columns = se_metastatic_df.columns.droplevel(0)

# Preview that Reformatting worked
metastatic_df.head()
```

```python
# Generate the Plot (with Error Bars)
capomulin2 = plt.errorbar(metastatic_df.index,
                         metastatic_df["Capomulin"],
                         yerr=se_metastatic_df["Capomulin"],
                         color='r',
                         marker='o',
                         linestyle='--')
infubinol2 = plt.errorbar(metastatic_df.index,
                         metastatic_df["Infubinol"],
                         yerr=se_metastatic_df["Infubinol"],
                         color='b',
                         marker='^',
                         linestyle='--')
ketapril2 = plt.errorbar(metastatic_df.index,
                         metastatic_df["Ketapril"],
                         yerr=se_metastatic_df["Ketapril"],
                         color='green',
                         marker='s',
                         linestyle='--')
placebo2 = plt.errorbar(metastatic_df.index,
                        metastatic_df["Placebo"],
                        yerr=se_metastatic_df["Placebo"],
                        color='black',
                        marker='d',
                        linestyle='--')

plt.grid()
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
plt.legend(handles=[capomulin2,infubinol2,ketapril2,placebo2])

# Save the Figure
plt.savefig("images/metastatic_spread_during_treatment.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count_grp = data_df[["Mouse ID","Timepoint","Drug"]].groupby(["Drug","Timepoint"])

# Convert to DataFrame
mice_count_df = mice_count_grp.count().rename(columns={"Mouse ID":"Mouse Count"})
# Preview DataFrame
 
mice_count_df.reset_index().head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_df = mice_count_df.unstack(level=0)
survival_df.columns = survival_df.columns.droplevel(0)

# Preview the Data Frame
survival_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
capomulin3 = plt.plot(survival_df.index,
                      survival_df["Capomulin"]/survival_df["Capomulin"].max()*100,
                      color='r',
                      marker='o',
                      linestyle='--')
infubinol3 = plt.plot(survival_df.index,
                      survival_df["Infubinol"]/survival_df["Infubinol"].max()*100,
                      color='b',
                      marker='^',
                      linestyle='--')
ketapril3 = plt.plot(survival_df.index,
                     survival_df["Ketapril"]/survival_df["Ketapril"].max()*100,
                     color='green',
                     marker='s',
                     linestyle='--')
placebo3 = plt.plot(survival_df.index,
                    survival_df["Placebo"]/survival_df["Placebo"].max()*100,
                    color='black',
                    marker='d',
                    linestyle='--')

plt.grid()
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend()

# Save the Figure
plt.savefig("images/survival_during_treatment.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
pct_list = []

for drug in tumor_df:
    initial = tumor_df[drug][0]
    final = tumor_df[drug][45]
    pct_list.append((final-initial)/initial*100)
                                                   
# Display the data to confirm
all_pct_ser = pd.Series(data=pct_list,index=tumor_df.columns)
all_pct_ser
```

```python
pct_ser = all_pct_ser.loc[["Capomulin","Infubinol","Ketapril","Placebo"]]

# create bar plot
x_axis = np.arange(len(pct_ser))

tumor_bar = plt.bar(x_axis, pct_ser, width=-1, align='edge', 
                    color=['g' if pct<0 else 'r' for pct in pct_ser])

plt.xticks([value for value in x_axis], pct_ser.index)
plt.yticks(np.arange(-20,70,20))
plt.ylim(-30,70)
plt.grid()

plt.title("Tumor Change Over 45 Day Treatment")
plt.xlabel("Drug")
plt.ylabel("Tumor Volume Change (%)")

# define label function
def label(rects,ax):
    
    for rect in rects:
        if rect.get_height() < 0:
            label_position = -8
        else:
            label_position = 5

        ax.text(rect.get_x() + rect.get_width()/2, label_position,
                f"{int(rect.get_height())}%",
                ha='center',
                color='w')

# call label function. rects = bar container obj returned from plt.bar, ax = plt.getcurrentaxes()
label(tumor_bar, plt.gca())

# Save the Figure
plt.savefig("images/tumor_change_over_45_day_treatment.png")

# Show the Figure
plt.show()
```

```python
Description of three observable trends based on the data:

1. Capomulin was the only drug that actually worked to decrease tumor volume by the end of the study. 
   The change in number of metastatic sites in mice treated with Capomulin was half of that in mice treated with the placebo,
   while tumor volume decreased by nearly 20%.
2. In mice treated with Infubinol, metastatic site spread was also drastically reduced, but tumor volume was only slightly reduced.
   Furthermore, the survival rate of mice in this group was lowest, even lower than that of mice treated with the placebo.
3. Mice treated with Ketapril actually faired worse than any other group.
   Both tumor volume and metastatic site spread were slightly increased in this group over the placebo group.
   However, the survival rate of mice in this group was essentially the same as that of mice in the placebo group.
```
